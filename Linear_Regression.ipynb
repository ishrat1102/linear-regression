```python
import csv
from array import *
```

```python
#--------------------DONE BY ISHRAT JABEN BUSHRA-------------
fields=[]
rows=[]
#rows is dataset array
Dataset=input("enter dataset:")
with  open(Dataset,'r') as csvfile:
    read=csv.reader(csvfile)
    next(read)
    for row in read:
            rows.append(row)
print(rows)
```

```python
for col in range(len(rows[0])):
    try:
        val = float(rows[0][col])
    except:
        cat = []
        cat.append(rows[0][col])
        for item in range(len(rows)):
            for st in range(len(cat)):
                if rows[item][col] == cat[st]:
                    rows[item][col] = st
                    break
                elif rows[item][col] not in cat:
                    cat.append(rows[item][col])
                    rows[item][col] = len(cat)-1
                    break
print(rows)
```

```python

col=len(rows[0])
ro=len(rows)
for i in range (ro):
    for j in range (col):
        rows[i][j]=float(rows[i][j])
print(rows)               
```

```python
col=len(rows[0])
ro=len(rows)
#print(col)
alpha=float(input("enter learning rate="))
w=[]
y=0
for i in range (col):
    weight=float(input("enter weights="))
    w.append(weight)

```

```python
err=[]
def regre():
    current=0
    temp=0
    w_new=[]
    error = []
    res=0.0
    summ=0.0
    for k in range (col):
        for i in range (ro):
            for j in range (col-1):
                if(j==0):
                    res+=w[j]
                    res+=w[j+1]*rows[i][j]
                else:
                    res+=w[j+1]*rows[i][j]
            if(k==0):
                error.append((res-rows[i][col-1])**2)
                summ += res-rows[i][col-1]       
            else:
                summ += (res -rows[i][col-1])*rows[i][k-1]
            res=0
        temp=w[k]-alpha*summ
        w_new.append(temp)
        temp = 0
        summ = 0

    err.append(sum(error)*0.5)
    print("Error is: " + str(err))
    #print(w_new)
    
    for i in range(len(w)):
        w[i] = w_new[i]
    

ite = int(input("enter iteration number="))    
print(w)

for i in range(ite):
    regre()
    if(i < ite-1 ):
       
        print(w) 
   
```
